```python
import pandas as pd
import re
import emoji

# Load your dataset (Only select relevant columns)
df = pd.read_csv("C:/Users/Prem Gupta/Downloads/data_to_be_cleansed.csv", usecols=["text", "label"])

# Function to clean text
def clean_text(text):
    if isinstance(text, str):  # Ensure the input is a string
        text = re.sub(r"http\S+|www\S+|https\S+", "", text)  # Remove URLs
        text = re.sub(r"@\w+", "", text)  # Remove mentions (@username)
        text = re.sub(r"#\w+", "", text)  # Remove hashtags
        text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove special characters
        text = emoji.replace_emoji(text, replace="")  # Remove emojis
        return text.lower().strip()
    return ""

# Apply cleaning function
df["cleaned_text"] = df["text"].apply(clean_text)

# Save cleaned data (without unnamed columns)
df.to_csv("cleaned_data_to_be_cleansed.csv", index=False)

print("✅ Data cleaning completed! Saved as cleaned_data.csv")
```
